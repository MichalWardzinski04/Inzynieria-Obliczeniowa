```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

A = list()
D = list()
E = list()
F = pd.DataFrame()
G = pd.DataFrame()
area = np.arange(0,np.pi/2,0.01)

for i in range(1,10001) : 
    x = np.random.rand(i,1)
    y = np.random.rand(i,1)
    a = x[x**2+y**2<1]
    b = y[x**2+y**2<1]
    c = x[x**2+y**2>=1]
    d = y[x**2+y**2>=1]
    A.append((len(a)*4/len(x)))
    
for i in [100,1000,10000] :
    for k in range(10):
        for l in range(10):
            x = np.random.rand(i,1)
            y = np.random.rand(i,1)
            e = x[x**2+y**2<1]
            E.append((len(e)*4/len(x)))    
        g = str(i) +'_' + str(k)
        G[g] = E
        E = list()

for i in range(10000,110000,10000) :
    for k in range(100):
        x = np.random.rand(i,1)
        y = np.random.rand(i,1)
        e = x[x**2+y**2<1]
        E.append((len(e)*4/len(x)))    
    g = str(i)
    F[g] = E
    E = list()

fig, axs = plt.subplots(2, 4, figsize=(12, 6))

ax1 = plt.subplot(2,4,1)         
ax1.scatter(c,d,color='blue',s=0.2)
ax1.scatter(a,b,color='red',s=0.2)
ax1.plot(np.cos(area),np.sin(area),color='black')
ax1.set_title("Losowe X Y")
ax1.set_ylim(0,1)
ax1.set_xlim(0,1)
ax1.set_xlabel("X")
ax1.set_ylabel("Y")

ax2 = plt.subplot(2,4,(2,4))         
ax2.plot(A,linewidth=0.2)
ax2.axhline(y=3.1466, color='black', linestyle='-')
ax2.set_title("Estymacja liczby PI 3.1466")
ax2.set_ylim(2,4)
ax2.set_xlim(0,10000)
ax2.set_xlabel("n")
ax2.set_ylabel("Estymacja pi")

for i in range(3):
    axs[1,i].boxplot(G.iloc[:,i*10:i*10+10],showfliers=False)
    axs[1,i].axhline(y=3.1466, color='red', linestyle='-')
    axs[1,i].set_ylim(2.5,3.5)
    axs[1,i].set_ylabel("Estymacja pi")
    axs[1,i].set_xlabel("Seria")
    axs[1,i].set_title(f'Po {100**(i+1)} losowaniach')
    
ax3 = plt.subplot(2,4,8)         
ax3.boxplot(F,showfliers=False)
ax3.set_title("Estymacja liczby PI 3.1466")
ax3.set_ylim(3.10,3.20)
ax3.set_xlabel("Liczba punktow Nx10^5")
ax3.set_ylabel("Estymacja pi")

fig.delaxes(axs[0,1])
fig.delaxes(axs[0,2])
fig.delaxes(axs[0,3])
plt.tight_layout()
plt.show()
```

```python
Projekt obrazuje dokladno przyblienia wartoci pi zalenego od iloci losowanych punktw w obszarze pierwszej wiartki koa.
Za pomoc pierwszej ptli wyznaczymy punkty znajdujace sie w okregu oraz poza nim co przedstawia pierwszy wykres.
Obok widzimy estymacje wartosci pi w zaleznosci od ilosci losowanych punktow.
W kolejnej ptli tworzymy 10 serii w ktorych losujemy zwikszajc si ilo punktw aby zobrazowa denie do wartoci pi.
Widac ze im wieksza ilo losowanych punktw to dokaldno sie zwieksza.
W ostatniej ptli liczymy przy tej samej iloci powtorze losowania i rnej iloci punktw warto pi.
Z wykresu atwo odczyta ze jest im wieksza liczba punktw tym lepsze przyblienie.
Wniosek jest taki ze ilo losowanych punktw ma duze znaczenie na wartoestymacji.
Dla najlepszego wyniku nalezy zmaksymalizowa ilo losowanych punktw
```
